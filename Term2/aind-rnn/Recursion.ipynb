# Coding up a recursive sequence

```python
def g(value, alpha, omega):
    return omega * value - omega * value**2
    

def recursive_sequence(alpha, omega, length):
    sequence = [alpha]
    for i in range(1, length):
        sequence.append(g(sequence[i-1], alpha, omega))
    return sequence
```

```python
seq1 = recursive_sequence(0.5, 1, 50)
seq2 = recursive_sequence(0.5, 3, 50)
seq3 = recursive_sequence(0.0001, 4, 50)
seq4 = recursive_sequence(1, 3, 50)
```
